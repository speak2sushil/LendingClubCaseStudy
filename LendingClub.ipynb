```python
# Calculate the threshold for dropping columns
threshold = 0.6 * len(lending)  # Keep columns with at least 60% non-null values
#print(threshold)

# Drop columns with more than 40% missing values
lending = lending.dropna(thresh=threshold, axis=1)
```

```python
lending.isnull().sum()*100/lending.shape[0]
```

```python


# **Drop unnecessary Columns**


lending= lending.drop(['desc'], axis=1)
lending= lending.drop(['emp_title'], axis=1)
lending= lending.drop(['title'], axis=1)
lending= lending.drop(['emp_length'], axis=1)
lending= lending.drop(['last_pymnt_d'], axis=1)
lending= lending.drop(['last_credit_pull_d'], axis=1)
lending= lending.drop(['pub_rec_bankruptcies'], axis=1)
lending.isnull().sum()*100/lending.shape[0]

```

```python

#Imputing 
```

```python
# Remove percentage symbols and convert 'revol_util' to numeric values
lending['revol_util'] = lending['revol_util'].str.rstrip('%').astype('float')

# # Calculate median as its numerical value after removing percentage symbols
med = lending['revol_util'].median()

# Fill NaN values with the calculated median
lending['revol_util'] = lending['revol_util'].fillna(med)
```

```python
# Convert int_rate column from string to float 
lending['int_rate'] = lending['int_rate'].str.rstrip('%').astype('float')
# Calculate correlation matrix
corr = lending[cont_cols].corr()


# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Inference
# From the output, we can see that 'total_rec_prncp', 'total_pymnt', 'total_pymnt_inv', and 'last_pymnt_amnt' have a relatively high positive correlation with the loan status. 
#This means that as these variables increase, the loan is more likely to be fully paid or current
```

```python
# Convert loan status to numerical values
lending['loan_status_num'] = lending['loan_status'].apply(lambda x: 1 if x in ['Fully Paid', 'Current'] else 0)


numerical_columns = lending.select_dtypes(include=['float64', 'int64']).columns

# Create a correlation matrix
correlation_matrix = lending[numerical_columns].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix[['loan_status_num']], annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation between loan_status_num and numerical columns')
plt.show()
```

```python
# The correlation coefficient ranges from -1 to 1. If the value is close to 1,
   #it means that there is a strong positive correlation between the two variables. 
    #When it is close to -1, the variables have a strong negative correlation.
# From the output, we can see that 'total_rec_prncp', 'total_pymnt', 'total_pymnt_inv', and 'last_pymnt_amnt' 
   #have a relatively high positive correlation with the loan status. 
    # This means that as these variables increase, the loan is more likely to be fully paid or current.
# On the other hand, 'loan_amnt', 'funded_amnt', 'dti', and 'recoveries' have a negative correlation with the loan status. 
  # This means that as these variables increase, the loan is less likely to be fully paid or current.
```

```python
# Bivariante Analysis for Categorical Column  
```

```python
for col in cat_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, hue='loan_status_num', data=lending, palette='coolwarm')
    plt.title(f'Count Plot: {col} vs Loan Status Num')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title='Loan Status Num', loc='upper right', labels=['Not Defaulted', 'Defaulted'])
    plt.show()
```

```python
# From the plots, we can see that:
# Most of the loans are 36 months term. The proportion of charged off or defaulted loans is higher for 60 months term loans.
# The grade of the loan also affects the loan status. Loans with high grades (B, C, A) have a higher proportion of being charged off or defaulted.
# The home ownership status of the borrower also influences the loan status. The proportion of charged off or defaulted loans is higher for borrowers who rent or own a home.
# The verification status of the borrower's income also affects the loan status. Loans where the income is not verified have a higher proportion of being charged off or defaulted.
```
